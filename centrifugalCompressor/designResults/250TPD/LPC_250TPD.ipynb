```python
import sys
import os
current_dir = os.path.dirname(os.path.abspath(''))
parent_dir = os.path.dirname(current_dir)
sys.path.append(parent_dir)
import compressor
```

```python
P_suction= 1.05663500000000 #bar
P_final  = 7.40142000000000 #bar

n_comp   = 3

R        = (P_final/P_suction)**(1/n_comp)

print(R)
```

```python
class Format:
    end = '\033[0m'
    underline = '\033[4m'

fluid   = "REFPROP::HYDROGEN"

#First low-pressure compressor
T_in    = 297.099916750707 #K
P_in    = 1.05663500000000 #bar
P_out   = R*P_in #bar

m       = round(4629.89033353505/3600,4) #kg/s

LPC1 = compressor.Compressor(fluid, T_in, P_in, P_out, m)
```

```python
LPC1.initialSpec()
```

```python
#design constraint
u2_max  = 500  #m/s

LPC1.noOfStages(u2_max)
```

```python
LPC1.rotationalSpeedRange()
```

```python
LPC1.nzSelectionPlot()
```

```python
LPC1.stagesCalculation(11200,8,True)
```
