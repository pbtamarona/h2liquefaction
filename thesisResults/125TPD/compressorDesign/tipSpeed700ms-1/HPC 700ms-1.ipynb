```python
P_suction= 7.40146000000000 #bar
P_final  = 29.8000000000000 #bar

n_comp   = 3

R        = (P_final/P_suction)**(1/n_comp)

print(R)
```

```python
from hydrogen_liquefaction import compressor

class Format:
    end = '\033[0m'
    underline = '\033[4m'

fluid   = "REFPROP::HYDROGEN"

#First high-pressure compressor
T_in    = 293.802876417080 #K
P_in    = 7.40146000000000  #bar
P_out   = R*P_in #bar

m       = round(51498.8305826687/3600,4) #kg/s

HPC1 = compressor.Compressor(fluid, T_in, P_in, P_out, m)
```

```python
HPC1.initialSpec()
```

```python
#design constraint
u2_max  = 700  #m/s

HPC1.noOfStages(u2_max)
```

```python
HPC1.rotationalSpeedRange()
```

```python
HPC1.nzSelectionPlot()
```
