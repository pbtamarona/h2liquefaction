```python
P_suction = 1.05913000000000 #bar
P_final   = 7.40146000000000 #bar

n_comp    = 3

R         = (P_final/P_suction)**(1/n_comp)

print(R)
```

```python
from hydrogen_liquefaction import compressor

class Format:
    end = '\033[0m'
    underline = '\033[4m'

fluid   = "REFPROP::HYDROGEN"

#First low-pressure compressor
T_in    = 297.099998889065 #K
P_in    = 1.05913000000000 #bar
P_out   = R*P_in #bar

m       = round(2315.23292650504/3600,4) #kg/s

LPC1 = compressor.Compressor(fluid, T_in, P_in, P_out, m)
```

```python
LPC1.initialSpec()
```

```python
#design constraint
u2_max  = 700  #m/s

LPC1.noOfStages(u2_max)
```
